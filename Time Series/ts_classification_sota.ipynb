```python
import pandas as pd

# Convert each row of X_train into a pandas Series and store them in a list
series_list = [pd.Series(row) for row in X_train]

# Create a pandas DataFrame from the list of Series
df_X_train = pd.DataFrame(series_list)

# Display the DataFrame
print(df_X_train)
```

### 1-NN with DTW (baseline)

```python
from sktime.classification.distance_based import KNeighborsTimeSeriesClassifier
```

```python
clf = KNeighborsTimeSeriesClassifier()
clf.fit(X_train, y_train)
```

```python
%%time
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

### ROCKET

```python
from sklearn.linear_model import RidgeClassifierCV
from sktime.transformations.panel.rocket import Rocket
```

```python
%%time
rocket = Rocket()  # by default, MiniRocket uses ~10,000 kernels
rocket.fit(X_train)
X_train_transform = rocket.transform(X_train)
clf = RidgeClassifierCV(alphas=np.logspace(-3, 3, 10)) #, normalize=True
clf.fit(X_train_transform, y_train)
```

```python
X_test_transform = rocket.transform(X_test)
y_pred = clf.predict(X_test_transform)
print(classification_report(y_test, y_pred))
```

### MINI-ROCKET

```python
from sklearn.linear_model import RidgeClassifierCV
from sktime.transformations.panel.rocket import MiniRocket
```

```python
%%time
minirocket = MiniRocket()  # by default, MiniRocket uses ~10,000 kernels
minirocket.fit(X_train)
X_train_transform = minirocket.transform(X_train)
clf = RidgeClassifierCV(alphas=np.logspace(-3, 3, 10), normalize=True)
clf.fit(X_train_transform, y_train)
```

```python
X_test_transform = minirocket.transform(X_test)
y_pred = clf.predict(X_test_transform)
print(classification_report(y_test, y_pred))
```

### Canonical Interval Forest (CIF)

```python
from sktime.classification.interval_based import CanonicalIntervalForest
```

```python
%%time
clf = CanonicalIntervalForest(n_estimators=30)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from tslearn.neighbors import KNeighborsTimeSeriesClassifier

# Set the Sakoe-Chiba radius (e.g., 5) to limit the search space for DTW computation
sakoe_chiba_radius = 5

# Create the classifier with the Sakoe-Chiba constraint
clf = KNeighborsTimeSeriesClassifier(metric='dtw', metric_params={'sakoe_chiba_radius': sakoe_chiba_radius})

# Train the classifier
clf.fit(X_train, y_train)
```

```python
%%time
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```
