# Elliptic Envelope

```python
from sklearn.covariance import EllipticEnvelope
```

```python
X = df_data.values
```

```python
clf = EllipticEnvelope(random_state=0, support_fraction=1)
clf.fit(X)
```

```python
outliers = clf.predict(X)
scores = abs(clf.score_samples(X))
```
